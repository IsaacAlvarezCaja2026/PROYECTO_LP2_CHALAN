```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración estética
sns.set_theme(style="whitegrid")

plt.rcParams["figure.figsize"] = (10,6)
plt.rcParams["font.size"] = 11
```

```python
df = pd.read_csv("precios_semana_emmsa.csv")

df["Fecha"] = pd.to_datetime(df["Fecha"], format="%d/%m/%Y")

df.head()
```

```python
producto="PAPA"

datos = (
    df[df["Producto"]==producto]
    .groupby("Fecha")["Prom"]
    .mean()
    .reset_index()
)

plt.figure(figsize=(10,5))

sns.lineplot(
    data=datos,
    x="Fecha",
    y="Prom",
    marker="o",
    linewidth=3
)

plt.title(f"Evolución del precio promedio de {producto}")

plt.xlabel("Fecha")
plt.ylabel("Precio promedio (S/.)")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(9,6))

sns.boxplot(
    data=df,
    x="Producto",
    y="Prom"
)

plt.title("Distribución de precios por producto")

plt.show()
```
